```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Banking.csv')
```

```python
df.head(5)
```

```python
#numerical columns in the data
df.describe()
```

```python
#Information of column names,number of rows and respective datatypes as well as non null count
df.info()
```

```python
#Check if there are missing values
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)
```

```python
#We will group the estimated incomes of customers
df.groupby('Estimated Income').count()

#create bins
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
#FRom above data we see that there are some columns which are categorical however turn out to be numerical
#Let us convert these columns into categorical
# Examine the distribution of unique cataegories in categorical columns
categorical_cols = df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns

for col in categorical_cols:
  print(f"Value Counts for '{col}':")
  display(df[col].value_counts())
```

```python
#So we have basically categorized the above columns into categories and counted the value_count() and using for loop through iterations for columns
#Categorical columns defined in list and accessed through indexing
```

### Univariate Analysis

```python
#Analysing categorical columns based on count

for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

```python
#for example we take amount of credit cards for univariate we just get the counts same if we check throigh other parameter we get more insights.
#If we want more insights we can use Bivariate analysis using hue

for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Nationality')
```

```python
# HIstplot of value counts for different Occupation

for col in categorical_cols:
  if col == "Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
```

##Numerical Analysis



```python

numerical_cols = ['Fee Structure','Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Univariate analysis and visualization
plt.figure(figsize=(25,20))
for i,col in enumerate(numerical_cols):   #enumerate used to fix the index with column names
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde=True)          #Kde=True Smooth line showing curve
  plt.title(col)                          #gives title of column names for each graph
plt.show()
```

##Heatmaps


```python
#We can see correlation betweeen numerical attributes
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()      #Correlation between numerical columns

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
#High correlation between Bank Deposits and Checking accounts.cx who maintain high balances in one acc type often hold substantial amt or funds
```

```python
#Find relationship between numerical columns through scatterplot

pairs_to_plot = [
    ('Bank Deposits', 'Saving Accounts'),
    ('Checking Accounts', 'Saving Accounts'),
    ('Checking Accounts', 'Foreign Currency Account'),
    ('Age', 'Superannuation Savings'),
    ('Estimated Income', 'Checking Accounts'),
    ('Bank Loans', 'Credit Card Balance'),
    ('Business Lending', 'Bank Loans'),
]


for x_col, y_col in pairs_to_plot:
    plt.figure(figsize=(8, 6))
    sns.regplot(                        #check regression line whether it is linear
        data=df,
        x=x_col,
        y=y_col,
        scatter_kws={'alpha': 0.4},     # semi-transparent points
        line_kws={'color': 'red'}       # best-fit line color
    )
    plt.title(f'Relationship between {x_col} and {y_col}', fontsize=14)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.tight_layout()
    plt.show()
```

##Insights

Together, these plots illustrate a financial lifecycle:

1.Income â†’ Liquidity (Estimated Income â†’ Checking Accounts)

2.Liquidity â†’ Savings Growth (Checking â†” Savings, Deposits â†” Savings)

3.Savings â†’ Diversification (Checking â†” Foreign Currency Accounts)

4.Credit Expansion (Business Lending â†” Bank Loans, Loans â†” Credit Cards)

5.Retirement Phase (Age â†” Superannuation Savings, showing drawdown)

##The scatter plot analysis reveals a coherent financial lifecycle: income drives liquidity, which supports savings growth and diversification. Credit expansion parallels this process, while retirement introduces a drawdown phase in superannuation savings. These relationships highlight the interconnected nature of personal and institutional financial behavior.
